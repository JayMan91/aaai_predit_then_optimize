```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
unweighted = pd.read_csv("Unweighted_QPTL.csv")
weighted = pd.read_csv("Weighted_QPTL.csv")
```

```python
fig, ax = plt.subplots(3, 3, figsize=(16,6))
ax1= ax[0,0]
ax2= ax[0,1]
ax3 = ax[0,2]
ax4 = ax[1,0]
ax5 = ax[1,1]
ax6 = ax[1,2]
ax7 = ax[2,0]
ax8 = ax[2,1]
ax9 = ax[2,2]
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==5],ax=ax1)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==10],ax=ax2)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==15],ax=ax3)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==20],ax=ax4)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==25],ax=ax5)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==30],ax=ax6)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==35],ax=ax7)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==40],ax=ax8)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= unweighted[unweighted.capacity==45],ax=ax9)
plt.show()
```

```python
fig, ax = plt.subplots(3, 3, figsize=(16,6))
ax1= ax[0,0]
ax2= ax[0,1]
ax3 = ax[0,2]
ax4 = ax[1,0]
ax5 = ax[1,1]
ax6 = ax[1,2]
ax7 = ax[2,0]
ax8 = ax[2,1]
ax9 = ax[2,2]
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= weighted[weighted.capacity==30],ax=ax1)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= weighted[weighted.capacity==60],ax=ax2)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= weighted[weighted.capacity==90],ax=ax3)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= weighted[weighted.capacity==120],ax=ax4)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= weighted[weighted.capacity==150],ax=ax5)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= weighted[weighted.capacity==180],ax=ax6)
g1 = sns.lineplot(x='subepoch', y='test_regret', legend="full",
              data= weighted[weighted.capacity==210],ax=ax7)

plt.show()
```
