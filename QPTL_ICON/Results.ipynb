```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
l1 =  pd.read_csv("Load1_qptl.csv")
l2 = pd.read_csv("Load2_qptl.csv")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,4))
ax1= ax[0]
ax2= ax[1]
g1 = sns.lineplot(x='subepoch', y='test_regret',markers=True, dashes=False,
                  legend="full",hue="hyperparams", style="hyperparams",
              data= l1,ax=ax1)
g2 = sns.lineplot(x='subepoch', y='test_regret',markers=True, dashes=False,
                  legend="full",hue="hyperparams", style="hyperparams",
              data= l2,ax=ax2)
plt.show()
```
